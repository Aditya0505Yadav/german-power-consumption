# MONTHLY Mean Power Consumption 2016

```python
pf=df.set_index("Date")
resampled=pf.resample("W").mean()  # can use "M"
resampled=resampled.reset_index()

resampled1=resampled[(resampled["Date"].dt.year==2016)]     #2017
resampled1.plot(x="Date",y="Consumption",color="green", marker= '*',linestyle='-',figsize=(5,3))
plt.title("monthly mean power consumption-2016")
plt.xlabel("Month")
plt.ylabel("consumption in GWH")
plt.grid()

resampled1.plot(x="Date",y="Solar",color="red", marker= '*',linestyle='-',figsize=(5,3))
plt.title("Monthly mean solar power consumption-2016")
plt.xlabel("Month")
plt.ylabel("Power Consumption")
plt.grid()

resampled1.plot(x="Date",y="Wind",color="blue", marker= '*',linestyle='-',figsize=(5,3))
plt.title("Monthly mean wind power consumption-2016")
plt.xlabel("Month")
plt.ylabel("Power Consumption")
plt.grid()
plt.show()
```

# MONTHLY AVERAGE POWER CONSUMPTION

```python
grouped=df.groupby(["Month","Year"]).sum()
a=[]
x=0
y=12
for i in range (0,12):
 a.append(grouped[x:y].Consumption.mean())
 x=y
 y=y+12
months=list(range(1,13))
months1=[]
months1=list(map(lambda x: calendar.month_abbr[x],months))
plt.subplots(figsize=(15,5))
plt.bar(months1,a)
plt.title("Monthly average consumption from 2006 to 2017")
plt.xlabel("Month")
plt.ylabel("Power Consumption")
plt.show()
```

# MONTHLY POWER CONSUMPTION 2006-17

```python
fig, axes = plt.subplots()
pf=df.set_index("Date")

resampled=pf.resample("M").sum()
resampled=resampled.reset_index()

resampled.plot(ax=axes,x="Date",y="Consumption")
resampled.plot(ax=axes,x="Date",y="Solar")
resampled.plot(ax=axes,x="Date",y="Wind")

plt.title("Monthly power consumption from 2006 TO 2017")
plt.xlabel("Year")
plt.ylabel("Power Consumption")
plt.grid()
plt.show()
```
